```python
import pandas as pd

df = pd.read_csv('/housing_dirty.csv')

df.head()
```

```python
df.info()

print(df.isnull().sum())

print("Duplikat:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)

df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()

df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].median())

print(df.isnull().sum())
```

```python
Q1 = df['harga_juta'].quantile(0.25)
Q3 = df['harga_juta'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['harga_juta'] >= lower) & (df['harga_juta'] <= upper)]

Q1 = df['luas_m2'].quantile(0.25)
Q3 = df['luas_m2'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['luas_m2'] >= lower) & (df['luas_m2'] <= upper)]

df = df[df['tahun_bangun'] >= 1950]

print("Missing Value:", df.isnull().sum().sum())
print("Duplikat:", df.duplicated().sum())
```
